Гипотеза:
Пиво предпочитают больше, чем вино или крепкие напитки, в странах где оценка счастья не меньше 7.

```python
data = pd.read_csv("/HappinessAlcoholConsumption.csv")
data.head(10)

# Country - Страна
# Region - Регион
# Hemisphere - Полушарие
# HapinessScore - Оценка счастья населения
# HDI - Индекс человеческого развития
# GDP_PerCapita - Индекс валового внутреннего продукта
# Beer_PerCapita - Литры выпитого пива
# Spirit_PerCapita - Литры выпитых крепких напитков
# Wine_PerCapita - Литры выпитого вина
```

```python
data['Country'].value_counts()
```

```python
count = 0
data = data[data['HappinessScore'] >= 7]
for country in data['Country']:
  sumbeer = int(data[data['Country'] == country]['Beer_PerCapita'])
  sumspirit = int(data[data['Country'] == country]['Spirit_PerCapita'])
  sumwine = int(data[data['Country'] == country]['Wine_PerCapita'])
  if sumbeer > sumwine and sumbeer > sumspirit:
    count += 1
print(f'Пиво предпочитают больше, чем другие алкогольные напитки в {count} странах из {len(data)}')
```

**Статистическая гипотеза оказалась верной**.  

```python
data2 = pd.read_csv("/HappinessAlcoholConsumption.csv")
countw, countb, counts = 0, 0, 0
for country in data2['Country']:
  sumbeer = int(data2[data2['Country'] == country]['Beer_PerCapita'])
  sumspirit = int(data2[data2['Country'] == country]['Spirit_PerCapita'])
  sumwine = int(data2[data2['Country'] == country]['Wine_PerCapita'])
  maxi = max(sumbeer, sumspirit, sumwine)
  if maxi == sumwine:
    countw += 1
  elif maxi == sumbeer:
    countb += 1
  elif maxi == sumspirit:
    counts += 1

plt.figure(figsize=(7, 7))
plt.title('Favourite alcohol in countries')
x = [countw, countb, counts]
labels = ['Wine', 'Beer', 'Spirit']

plt.pie(x, labels=labels);
```
